# CityBikes

#### Send a request to CityBikes for the city of your choice. 

```python
import requests
import os
# Get request for citybike in Montreal, QC, Canada

citybike_url='http://api.citybik.es/v2/networks/bixi-montreal'

headers = {"Accept": "application/json"}

citybike_result = requests.get(citybike_url, headers=headers)
print(citybike_result)

# Read the result from the response of the get request 
data_load_1=citybike_result.text
```

#### Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# Load the response as json file
import json
data_1 = json.loads(data_load_1)
data_1
```

#### Put your parsed results into a DataFrame.

```python
# Extract stations data from network and turn it into a dataframe
import pandas as pd
raw_data_1= pd.json_normalize(data_1["network"]["stations"])
raw_data_1
```

#### Finalize dataframe
- Create a dataframe for analysis for the project with station names, bikes available, longitude and latitude

```python
# Create a dictionary with the column of dataframe as the key
citybike_dict= {'Station_name':[],
                 'Available_bikes': [],
                 'Longitude':[],
                 'Latitude':[]}

# Insert the data from the raw_data into citybike_dict
citybike_dict['Station_name']   = raw_data_1['name'].astype(object) 
citybike_dict['Available_bikes']= raw_data_1['free_bikes'].astype(int)
citybike_dict['Longitude']      = raw_data_1['longitude'].astype(float)
citybike_dict['Latitude']       = raw_data_1['latitude'].astype(float)
```

```python
# Put the dict into a dataframe
citybike_df=pd.DataFrame(citybike_dict)

citybike_df
```
